```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lorentz(nu, nu0, Gamma):
    return 1/np.pi * Gamma/2 * 1/(np.power((nu-nu0),2) + Gamma**2/4)

num_eigs = 100
L = 10e-6
vs = 18e3
nu_x = []
nu_y = []
nu_z = []
nu = np.linspace(0,vs/(2*L)*num_eigs, 1000)
rho = np.empty_like(nu)

for j in range(3):
    for i in range(num_eigs):
        rho += lorentz(nu, vs/(2*L)*i, 0.1*vs/(2*L)*i)

```
