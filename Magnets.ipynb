```python
import numpy as np
from numpy import pi as pi
from numpy import sqrt as sqrt
import matplotlib.pyplot as plt
```

```python
mu0 = 4*pi*1e-7
```

```python
def helmholtz(I, R, n):
    return (4/5)**1.5*n*mu0*I/R

def Length(n, r, R):
    return 2*pi*(R+r)*n

def Resistance(rho, L, r):
    return rho/(pi*r**2)*L
```

```python
rho = 0.017*(1e-3)**2

I = 0.1
n = 500
r = 50e-6
R = 5e-3
Rs = np.linspace(1e-3, 10e-3, 100)

fig = plt.figure()
ax1 = fig.gca()
ax2 = ax1.twinx()
ax1.plot(Rs, helmholtz(I, Rs, n)*1e3)
ax2.plot(Rs, Resistance(rho, Length(n, r, Rs), r) * I**2)
# print("Field strength: ", helmholtz(I, R, n))
# print("Power dissipation: ", Resistance(rho, Length(n, r, R), r) * I**2)
print("Form factor: ", np.sqrt(n)*2*r)
```
