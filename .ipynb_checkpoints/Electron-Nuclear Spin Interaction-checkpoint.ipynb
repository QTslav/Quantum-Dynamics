```python
from qutip import *
import numpy as np
from numpy import pi as pi
from numpy import sqrt as sqrt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot, animation
import sympy
import scipy.linalg as sl
```

```python
def pulse(t, t0, width):
    return np.where( (t0<t) & (t<(width+t0)), 1, 0)
```

```python
Su = basis(2,0)
Sd = basis(2,1)
Iu = basis(2,0)
Id = basis(2,1)

Sx = sigmax()
Sy = sigmay()
Sz = sigmaz()
Sm, Sp = Sx.eigenstates()[1]

Ix = sigmax()
Iy = sigmay()
Iz = sigmaz()
Im, Ip = Ix.eigenstates()[1]

omega_L_S = 1
omega_L_I = 1
A_perp = 0
A_para = 0
H0 = omega_L_S/2 * tensor(Sz, qeye(2)) + omega_L_I/2 * tensor(qeye(2), Iz)
HHF = A_para * tensor(Sz, Iz) + A_perp * tensor(Sz, Ix)
HSys = H0 + HHF

Omega_pi = 2
T = pi/Omega_pi
tau = 2*pi*1/omega_L_S
N = 16 
t = np.arange(0, 1*(2*T + 4*tau), T/50)

Hpi = Omega_pi/2*tensor(Sy, qeye(2))
Htau = HSys
pi_seq =  pulse(t, tau, T) + pulse(t, 3*tau+T, T)
tau_seq = np.where(pi_seq==1, 0, 1)
pi_seqs = pi_seq.copy()
tau_seqs = tau_seq.copy()
for i in range(0, N):
    pi_seqs = np.append(pi_seqs, pi_seq)
    tau_seqs = np.append(tau_seqs, tau_seq)
    
fig, axs = plt.subplots(3, 1)
axs[0].plot(t, pi_seq, alpha=0.3, color='k', ls='-.', lw=2)
axs[1].plot(t, tau_seq, alpha=0.3, color='r', ls='--', lw=2)

rho0 = tensor(Sp*Sp.dag(), Iu*Iu.dag())
output = mesolve([[HSys, tau_seq], [Hpi, pi_seq]], rho0, t, e_ops = [tensor(Sx, qeye(2)), tensor(Sy.dag(), qeye(2)), tensor(Sz.dag(), qeye(2))])

for exp in output.expect:
    axs[2].plot(t, exp, ls='-', lw=2)
    axs[2].fill_between(t, 0, pi_seq, color='k', alpha=0.2)    
    axs[2].fill_between(t, 0, tau_seq, color='r', alpha=0.2)

plt.tight_layout()
```

```python
import time
fig = plt.figure()
ax = Axes3D(fig, azim=-40, elev=30)
sphere = qutip.Bloch(axes=ax)

def animate(i):
    sphere.clear()
    sphere.add_vectors([output.expect[0][i], output.expect[1][i], output.expect[2][i]])
    sphere.make_sphere()
    return ax

t = time.time()
ani = animation.FuncAnimation(fig, animate, np.arange(len(output.expect[0])), blit=False, repeat=False)
ani.save('bloch_sphere.gif', fps=60)
print(time.time() - t)
```

```python
mu_I = 10.7e6 * 2*pi
B = 293e-4
w_L = mu_I * B
w_h = 2*pi*40e3
theta = pi/4.5
N = 32

def phi(tau, w_L, w_h, theta):
    A = w_h*np.cos(theta)
    B = w_h*np.sin(theta)    
    w_tilde = sqrt((A+w_L)**2 + B**2)
    m_x = B/w_tilde
    m_z = (A+w_L)/w_tilde
    
    alpha = w_tilde*tau    
    beta = w_L*tau
    
    phi = np.arccos(np.cos(alpha)*np.cos(beta) - m_z*np.sin(alpha)*np.sin(beta))
    return phi

def n1n2(tau, w_L, w_h, theta):
    A = w_h*np.cos(theta)
    B = w_h*np.sin(theta)    
    w_tilde = sqrt((A+w_L)**2 + B**2)
    m_x = B/w_tilde
    m_z = (A+w_L)/w_tilde
    
    alpha = w_tilde*tau    
    beta = w_L*tau
    
    return -m_x**2*(1-np.cos(alpha))*(1-np.cos(beta))/(1 + np.cos(alpha)*np.cos(beta) - m_z*np.sin(alpha)*np.sin(beta)) + 1

def M(n1n2_, phi_, N):
    return 1-(1-n1n2_)*np.sin(N*phi_/2)**2

def Px(M_):
    return (M_+1)/2

taus = np.arange(1, 100, 0.02)*1e-6
phi_ = phi(taus, w_L, w_h, theta)
n1n2_ = n1n2(taus, w_L, w_h, theta)
M_ = M(n1n2_, phi_, N)
Px_ = Px(M_)
plt.plot(taus, Px_, '-r')

# phi_ = phi(taus, w_L, 2*w_h, theta/2)
# n1n2_ = n1n2(taus, w_L, 2*w_h, theta/2)
# M_ = M(n1n2_, phi_, N)
# Px_ = Px(M_)
# plt.plot(taus, Px_, '-b')
```
